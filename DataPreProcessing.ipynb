# Data Visualization

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('data/penguins_raw.csv')
```

### Data Investigation: Initial Overview

Let's start by displaying the first few rows of the DataFrame to get a glimpse of the data, its columns, and the format.

```python
display(df.head())
```

Next, we'll check the DataFrame's information, which includes the column names, non-null counts, and data types. This helps us understand the structure and identify potential issues like incorrect data types or columns with many missing values.

```python
print(df.info())
```

Finally, let's count the number of missing values for each column. This is crucial for deciding how to handle missing data during preprocessing.

```python
print(df.isnull().sum())
```

Based on the high number of missing values, the 'Comments' column is not very useful and can be dropped.

```python
# Drop the 'Comments' column due to a high number of missing values
df = df.drop('Comments', axis=1)

# Display the number of missing values after dropping 'Comments'
print(df.isnull().sum())
```

Now, let's address the remaining missing values.

For numerical features like 'Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)', 'Delta 15 N (o/oo)', and 'Delta 13 C (o/oo)', we can impute the missing values with the mean or median. Given the small number of missing values for some of these, the mean is a reasonable choice.

For the categorical feature 'Sex', we will impute the missing values with the mode (the most frequent value).

```python
# Impute numerical columns with the mean
numerical_cols = ['Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)', 'Delta 15 N (o/oo)', 'Delta 13 C (o/oo)']
for col in numerical_cols:
    df[col].fillna(df[col].mean(), inplace=True)

# Impute 'Sex' column with the mode
df['Sex'].fillna(df['Sex'].mode()[0], inplace=True)

# Display the number of missing values after imputation
print(df.isnull().sum())
```

Now that we've handled all missing values, the dataset is cleaner. Let's confirm the data types and look at some summary statistics.

```python
print(df.info())
display(df.describe())
```

We must also check for duplicated values (if there exist any).

```python
print("Number of duplicates:", df.duplicated().sum())
```

Now that missing values are handled and we have an overview of the data types, let's investigate the unique values in the categorical columns. This will help us understand the distinct categories and identify potential areas for filtering or aggregation.

```python
# Identify categorical columns (object type)
categorical_cols = df.select_dtypes(include='object').columns

print("Unique values and their counts for categorical columns:")
for col in categorical_cols:
    print(f"\n--- Column: {col} ---")
    print(df[col].value_counts())

# Also check the 'Date Egg' column, which appears to be an object type but represents dates.
# It should ideally be converted to a datetime object for any time-series analysis or filtering by date.
print("\n--- Column: Date Egg ---")
print(df['Date Egg'].value_counts())
```

Based on the unique values:

*   **'Species'**: This is a key categorical variable that will likely be used for grouping or analysis. We might filter for specific species or aggregate metrics by species.
*   **'Region'**, **'Island'**: These also look like good candidates for grouping or filtering, depending on the analysis goals.
*   **'Stage'**, **'Clutch Completion'**, **'Sex'**: These are fairly clean and can be used directly or for filtering.
*   **'studyName'**, **'Individual ID'**: These columns have many unique values, almost one per row. 'Individual ID' is an identifier and is likely not useful for direct analysis or aggregation unless we need to track specific individuals. 'studyName' has only a few unique values, so it could potentially be used for grouping or filtering if there are differences across studies.
*   **'Date Egg'**: This column is currently an `object` type. It should be converted to a `datetime` object to enable time-based filtering or aggregation (e.g., daily, monthly counts).

```python
# Convert 'Date Egg' to datetime
df['Date Egg'] = pd.to_datetime(df['Date Egg'])

# Verify the conversion
print("\n--- After converting 'Date Egg' to datetime ---")
print(df['Date Egg'].dtype)
print(df.info())
```

At this point, the need for further filtering or aggregation depends entirely on the specific questions we want to answer. We've cleaned up missing values and ensured `Date Egg` is in the correct format.

**Possible next steps for filtering/aggregation could include:**

1.  **Filtering by Species or Island**: If we want to analyze a specific group of penguins.
2.  **Aggregating by Species, Island, or Date**: To see average culmen lengths, body masses, etc., for different groups or over time.
3.  **Feature Engineering**: Extracting year, month, or day from 'Date Egg' for time-series analysis.

To illustrate the process of aggregation, we will calculate the average values ​​of key physical attributes (beak length, beak depth, fin length, and body mass) for each penguin species on each island. This helps us better understand the differences between species and the influence of habitat.

```python
# Aggregation Example: Calculate mean physical attributes by Species and Island
agg_df = df.groupby(['Species', 'Island'])[['Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)']].mean().reset_index()

print("Mean physical attributes by Species and Island:")
display(agg_df)

# Another example: Count the number of penguins by Species and Island
count_df = df.groupby(['Species', 'Island']).size().reset_index(name='Count')
print("\nNumber of penguins by Species and Island:")
display(count_df)
```

Let's explore other aggregation techniques to gain a more comprehensive view of the data, including descriptive statistics such as minimum, maximum, median, and standard deviation, as well as counting the number of unique values.

```python
# Aggregation Example: Calculate min, max, median, and std deviation of physical attributes by Species and Island
agg_stats_df = df.groupby(['Species', 'Island'])[['Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)']].agg(['min', 'max', 'median', 'std']).reset_index()

print("Min, Max, Median, and Std Deviation of physical attributes by Species and Island:")
display(agg_stats_df)
```

Counting the number of unique values ​​('Individual ID') within each Species and Island group can help us understand the diversity of the recorded individuals.

```python
# Aggregation Example: Count unique Individual IDs by Species and Island
unique_id_count_df = df.groupby(['Species', 'Island'])['Individual ID'].nunique().reset_index(name='Unique Individual IDs')

print("Number of unique Individual IDs by Species and Island:")
display(unique_id_count_df)
```

You can filter the DataFrame to include only records of a specific penguin species.

```python
# Filtering Example: Filter by a specific Species
adelie_penguins = df[df['Species'] == 'Adelie Penguin (Pygoscelis adeliae)']

print("Data for Adelie Penguins:")
display(adelie_penguins.head())
```

Additionally, we can filter the data based on an arithmetic condition, such as 'Culmen Length (mm)' being greater than 50mm. This allows us to focus on penguins with longer beaks.

```python
# Filtering Example: Filter by a specific Island
gentoo_penguins = df[df['Species'] == 'Gentoo penguin (Pygoscelis papua)']

print("Data for Gentoo Penguins:")
display(gentoo_penguins.head())

# Filtering Example: Filter by a numerical condition (e.g., Culmen Length > 50mm)
long_culmen_penguins = df[df['Culmen Length (mm)'] > 50]

print("\nPenguins with Culmen Length > 50mm:")
display(long_culmen_penguins.head())
```

Finally, we can combine multiple conditions to filter more complex data. For example, we can search for female penguins of the species 'Adelie Penguin (Pygoscelis adeliae)' and from 'Biscoe' Island.

```python
# Filtering Example: Filter by multiple conditions (e.g., Species and Sex and Island)
ad_fem_biscoe_penguins = df[(df['Species'] == 'Adelie Penguin (Pygoscelis adeliae)') &
                            (df['Sex'] == 'FEMALE') &
                            (df['Island'] == 'Biscoe')]

print("\nFemale Adelie Penguins from Biscoe Island:")
display(ad_fem_biscoe_penguins.head())
```

We can use isin() to filter for complex conditions

```python
# Filtering Example: Filter by multiple species using isin()
selected_species = ['Adelie Penguin (Pygoscelis adeliae)', 'Gentoo penguin (Pygoscelis papua)']
multi_species_penguins = df[df['Species'].isin(selected_species)]

print("\nPenguins from Adelie or Gentoo species:")
display(multi_species_penguins.head())
```
